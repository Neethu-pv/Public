# Handling null values

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'median')
data[numerical_features] = imputer.fit_transform(data[numerical_features])
```

```python
data.isna().sum()
```

```python
# Handle missing values for categorical columns
data['gender'].fillna(data['gender'].mode()[0], inplace=True)
data['strong_foot'].fillna(data['strong_foot'].mode()[0], inplace=True)
data['coaching'].fillna(data['coaching'].mode()[0], inplace=True)

# Convert weight to numeric (extract numeric part and convert)
data['weight'] = data['weight'].str.extract('(\d+)').astype(float)

# Handle missing values for numerical columns
data['weight'].fillna(data['weight'].mean(), inplace=True)
```

```python
data.isna().sum()
```

# Handling Outlier

```python
def handle_outliers_iqr(df):
    for column in df.columns:
        if df[column].dtype != 'O':  # Only apply to numerical columns
            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
            df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])
    return df
```

```python
data = handle_outliers_iqr(data)
```
