# handling of outlier in case of the column age

```python
Q1 = np.percentile(data['Age'],25,method='midpoint')
Q2 = np.percentile(data['Age'],50,method='midpoint')
Q3 = np.percentile(data['Age'],75,method='midpoint')
```

```python
print("Q1 is ",Q1)
print("Q2 is ",Q2)
print("Q3 is ",Q3)
```

```python
IQR=Q3-Q1
IQR
```

```python
low_lim = Q1 - 1.5 * IQR
up_lim = Q3 + 1.5 * IQR

print("lower limit is ",low_lim)
print("upper limit is ",up_lim)
```

```python
outlier=[]
for x in data['Age']:
  if((x>up_lim)or(x<low_lim)):
   outlier.append(x)

outlier
```

```python
data['Age']=data['Age'].clip(lower=low_lim,upper=up_lim)
```

```python
outlier=[]
for x in data['Age']:
  if((x > up_lim) or (x < low_lim)):
    outlier.append(x)

outlier
```

```python
plt.boxplot(data['Age'])
plt.show()
```
